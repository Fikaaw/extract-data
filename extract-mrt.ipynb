Describe Data:

`nid` = id stasiun

`title` = nama stasiun

`urutan` = mrt adalah transportasi yang mmiliki satu urutan rute dari Stasiun Lebak Bulus ke Stasiun Bundaran HI dan sebaliknya dengan melewati 11 stasiun.

`isbig` = tidak diketahui dan juga tidak menandakan apakah stasiun itu besar atau kecil karena stasiun Bundaran HI termasuk besar namun dalam kategori 0.

`path` = menunjukkan path atau jalur dalam mengambil data.

`catatan` = akan di drop karena tidak menjelaskan apa-apa

`Antarmodas` = Integrasi, ketersediaan layanan lanjutan, serta akses antarmoda merupakan fasilitas yang penting dalam mobilitas keseharian masyarakat selain integrasi sistem pembayaran serta koridor layanan antarmoda tersebut. Sepanjang koridor Lebak Bulus hingga Bundaran HI, MRT Jakarta bekerja sama dengan operator transportasi publik lainnya seperti BRT Transjakarta untuk menyediakan integrasi tersebut, mulai dari integrasi fisik langsung seperti yang terlihat di stasiun dan halte Bundaran HI serta stasiun dan halte Lebak Bulus.

`jadwal_lb_biasa` =

none-> tidak ada jadwal biasa untuk stasiun lebak bulus ke lebak bulus.

`jadwal_lb_libur` =

none-> tidak ada jadwal libur untuk stasiun lebak bulus ke lebak bulus.

Hari Biasa: Jadwal kereta MRT dimulai pada pukul 05:15, dengan kereta terakhir pukul 23:40 WIB.

Hari Libur: Jadwal kereta MRT dimulai pada pukul 05:15, dengan kereta terakhir pukul 23:40 WIB.


`jadwal_hi_biasa` =

none-> tidak ada jadwal biasa untuk stasiun bundaran hi ke stasiun bundaran hi.

`jadwal_hi_libur` =

none-> tidak ada jadwal libur untuk stasiun bundaran hi ke stasiun bundaran hi.

`retails` =
data retail di tiap stasiun, meliputi

`estimasi` =
estimasi waktu dari satu stasiun ke stasiun lain.

`fasilitas` =
fasilitas di tiap stasiun.
